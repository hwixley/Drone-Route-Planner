```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import loadtxt
fileNames = ["GST","GTS","GT","GS","TS","ST","G","T","S"]
names = ["Greedy -> Swap -> 2-Opt", "Greedy -> 2-Opt -> Swap", "Greedy -> 2-Opt", "Greedy -> Swap",
         "2-Opt -> Swap","Swap -> 2-Opt","Greedy","2-Opt","Swap"]
colors = ["red","darkorange","green","blue","blue","deeppink","purple","red","cyan"]

moves = [[],[],[],[],[],[],[],[],[]]
for i in range(8):
    moves[i] = list(map(int,loadtxt("../aqmaps"+fileNames[i]+"Moves.txt",dtype="str",delimiter="\n",unpack=False)))

dates = loadtxt("../aqmapsDates.txt",dtype="str",delimiter="\n",unpack=False)
```

```python
import seaborn as sns
import pandas as pd
x = range(92)

plt.rcParams["figure.figsize"] = (13,8)

for i in range(7):
    if i != 3:
        plt.scatter(x,moves[i][0:92],c=colors[i],label=names[i],alpha=0.75)
        plt.plot(x, np.poly1d(np.polyfit(x,moves[i][0:92],1))(np.unique(x)),c=colors[i],lw=3)

plt.grid(True)
plt.legend(loc="lower left")
plt.xlabel("Day #",fontsize=16)
plt.ylabel("# Moves",fontsize=16)
plt.xticks(x,range(1,93),rotation=90,fontsize=7)
plt.title("A graph to show route performance of different high\ncomplexity algorithms for all unique days",fontsize=20)
```

```python
ys=[3,7]

for y in ys:
    plt.scatter(x,moves[y][0:92],c=colors[y],label=names[y])
    plt.plot(x, np.poly1d(np.polyfit(x,moves[y][0:92],1))(np.unique(x)),c=colors[y],lw=3)

plt.grid(True)
plt.legend(loc="lower left")
plt.xlabel("Day #",fontsize=16)
plt.ylabel("# Moves",fontsize=16)
plt.xticks(x,range(1,93),rotation=90,fontsize=7)
plt.title("A graph to show route performance of different low\ncomplexity algorithms for all unique days",fontsize=20)
```
